```python
#here we will tokenize our dataset
tokens=word_tokenize(Document)
print(tokens)
```

```python
#build vocabulary
vocab={ '<unkn>':0 }
#the counter counts how many times a word is repeated
# Counter(tokens).keys()
for token in Counter(tokens).keys():
  if token not in vocab:
    vocab[token]=len(vocab)
print(vocab)
```

```python
len(vocab)
```

```python
# now we need to extract the sentances from the document , for this we use the split function
input_sentences=Document.split('.')
print(input_sentences)
#lets split each word from the sentances
for sentence in input_sentences:
  print(word_tokenize(sentence))
#for each of these tokens we have to replace with the number in the vocab
```

```python
# to do so we need make a function to convert it into numericalsentence
def text_to_indices(sentence,vocab):
  numericalsentence=[]
  for token in sentence:
    if token in vocab:
      numericalsentence.append(vocab[token])
    else :
      numericalsentence.append(vocab['<unkn>'])
  return numericalsentence
```

```python
input_num_sen=[]
for sentence in input_sentences:
  #here we have successfully converted the sentence to numerical sentence ,
  # print(text_to_indices(word_tokenize(sentence),vocab))
  #after converting we need to make a list of these numerical sentences
  input_num_sen.append(text_to_indices(word_tokenize(sentence),vocab))
```

```python
input_num_sen
```

```python
#now we need to convert these numerical sentences into learnable parameters
#suppose a sentence[1,2,3] we can generate sequences like [1,2]->[1,2,3],,,lets code this
training_seq=[]
for sentence in input_num_sen:
   for i in range(1,len(sentence)):
      training_seq.append(sentence[:i+1])

```

```python
#we have made a training sequence ,but the nmerical sentences in training sequence have different lengths and we cannot train LSTM over this
#we need to make size of every sentence same ,using padding i.e [1,2]->[0,1,2],to do that we need to find longest sequence
len_list=[]
for i in range (0,len(training_seq)):
   len_list.append(len(training_seq[i]))
max(len_list)
```

```python
#here we add zeroes to the seq in training seq
padded_training_seq=[]
for seq in training_seq:
    padded_training_seq.append([0]*(max(len_list)-len(seq))+seq)


```

```python
#now we will make a tensor of this padded seq
padded=torch.tensor(padded_training_seq)
padded
```

```python
#now lets create X and Y for training
X=padded[:,:-1] #for all rows and upto second last element
Y=padded[:,-1]

```

```python
class CustomDataset(Dataset):
  def __init__(self,X,Y):
      super().__init__() #call the dataset's constructor first,then run mine
      self.X = X
      self.Y = Y

  def __len__(self):
    return self.X.shape[0] # Corrected: shape is an attribute, not a method; access first dimension

  def __getitem__(self, idx): # Corrected: __getitem__ needs an index argument
    return self.X[idx],self.Y[idx]
```

```python
dataset=CustomDataset(X,Y)
# type(dataset)
len(dataset)
dataset[0]
```

```python
dataloader=DataLoader(dataset,batch_size=32,shuffle=True)
```

```python
for input ,output in dataloader:
  print(input,output)
```

```python
#now we will make our LSTM model
class LSTMModule(nn.Module):
  def __init__(self,vocab_size):
    super().__init__()
    self.embedding=nn.Embedding(vocab_size,100)
    self.lstm=nn.LSTM(100,150,batch_first=True)
    self.fc=nn.Linear(150,vocab_size)

  def forward(self,x):
    embeded=self.embedding(x)
    # nn.lstm(embeded)
    intermediate_hidden_state,(final_hidden_state,final_cell_state)=self.lstm(embeded)
    output=self.fc(final_hidden_state.squeeze(0))
    return output

```

```python

model=LSTMModule(len(vocab))
```

```python
#lets run take our model to GPU
device=torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
epochs=50
learning_rate=0.001
criterion=nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(model.parameters(),lr=learning_rate)
```

```python
#training loop
for epoch in range(epochs):
  total_loss =0
  for batch_x,batch_y in dataloader:
    batch_x,batch_y=batch_x.to(device),batch_y.to(device)
    optimizer.zero_grad()
    output=model(batch_x)
    loss=criterion(output,batch_y)
    loss.backward()
    optimizer.step()
    total_loss+=loss.item()
  print(f"Epoch:{epoch+1},Loss:{total_loss:.4f}")
```

```python
#prediction
def prediction(model,vocab,text):
  #tokenize
  tokenized_text=word_tokenize(text.lower())
  # print(tokenized_text)

  #text ->indices
  numerical_text=text_to_indices(tokenized_text,vocab)
  # print(numerical_text)

  #padding (as we may don't know the size of the text sentence provided by the user)
  padded_text=torch.tensor([0]*(max(len_list)-len(numerical_text))+numerical_text,dtype=torch.long).unsqueeze(0)
  # print(padded_text)

  #padded -> model
  output=model(padded_text)
  # print(output)

  #we findout the index of the pridicted word
  value,index=torch.max(output,dim=1)
  # print(value,index)
  # print(list(vocab.keys())[index])
  #fetch that word and merge it with the sentance
  return text + " " +list(vocab.keys())[index]

```
